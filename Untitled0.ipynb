```python

# Ввод A0
print("Введите значение А0 (допустимые значение от -10 В до 10 В)")
A0 = float(input())
while A0 < -10 or A0 > 10:
  print("Некорректное значение, введите значение А0 ещё раз.")
  A0 = float(input())
# Ввод А1
print("Введите значение А1 (допустимые значения от 0 В до 220 В)")
A1 = float(input())
while A1 <0 or A1 > 220:
  print("Некорректное значение, введите значение A1 ещё раз.")
  A1 = float(input())
# Ввод А2
print("Введите значение А2 (допустимые значения от 0 В до 220 В)")
A2 = float(input())
while A2 <0 or A2 > 220:
  print("Некорректное значение, введите значение А2 ещё раз.")
  A2 = float(input())
# Ввод А3
print("Введите значение А3 (допустимые значения от 0 В до 220 В)")
A3 = float(input())
while A3 <0 or A3 > 220:
  print("Некорректное значение, введите значение А3 ещё раз.")
  A3 = float(input())
# Ввод f1
print("Введите значение f1 (допустимые значения: 50, 100, 150, 200, 250, 300, 350, 400, 450, 500 Гц)")
f1 = int(input())
while f1 != 50 and f1 != 100 and f1 != 150 and f1 != 200 and f1 != 250 and f1 != 300 and f1 != 350 and f1 != 400 and f1 != 450 and f1 != 500:
  print("Некорректное значение, введите значение f1 ещё раз.")
  f1 = int(input())
# Ввод f2
print("Введите значение f2 (допустимые значения: 50, 100, 150, 200, 250, 300, 350, 400, 450, 500 Гц)")
f2 = int(input())
while f2 != 50 and f2 != 100 and f2 != 150 and f2 != 200 and f2 != 250 and f2 != 300 and f2 != 350 and f2 != 400 and f2 != 450 and f2 != 500:
  print("Некорректное значение, введите значение f2 ещё раз.")
  f2 = int(input())
# Ввод f3
print("Введите значение f3 (допустимые значения: 50, 100, 150, 200, 250, 300, 350, 400, 450, 500 Гц)")
f3 = int(input())
while f3 != 50 and f3 != 100 and f3 != 150 and f3 != 200 and f3 != 250 and f3 != 300 and f3 != 350 and f3 != 400 and f3 != 450 and f3 != 500:
  print("Некорректное значение, введите значение f3 ещё раз.")
  f3 = int(input())
# Ввод W1
print("Введите значение W1 (допустимые значения от 0 до 360 градусов)")
W1 = float(input())
while W1 < 0 or W1 > 360:
  print("Некорректное значение, введите значение W1 ещё раз.")
  W1 = float(input())
# Ввод W2
print("Введите значение W2 (допустимые значения от 0 до 360 градусов)")
W2 = float(input())
while W2 < 0 or W2 > 360:
  print("Некорректное значение, введите значение W2 ещё раз.")
  W2 = float(input())
# Ввод W3
print("Введите значение W3 (допустимые значения от 0 до 360 градусов)")
W3 = float(input())
while W3 < 0 or W3 > 360:
  print("Некорректное значение, введите значение W3 ещё раз.")
  W3 = float(input())
# Ввод шага по времени
print("Введите значение шага по времени")
time = float(input())
# Импорт библиотек
import matplotlib.pyplot as plt
import numpy as np
import math as m
# Конвертируем значение углов из градусов в радианы
WR1 = m.radians(W1)
WR2 = m.radians(W2)
WR3 = m.radians(W3)
# Задание диапазона и функции
t = np.arange(0, 2*np.pi, time)
S = A0 + A1*np.sin(2*np.pi*f1*t + WR1) + A2*np.sin(2*np.pi*f2*t + WR2) + A3*np.sin(2*np.pi*f3*t + WR3)
# Построение графика
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(t, S, 'black', lw=2, label = "S=A0+A1sin(2pif1t+W1)+A2sin(2pif2t+W2)+A3sin(2pif3t+W3)")
ax.spines['bottom'].set_color('blue')     # Верхняя ось
ax.spines['left'].set_color('red')  # Правая ось
# Максимальное, минимальное, среднее значение напряжения
max_point = np.max(S)
plt.plot(t[np.argmax(S)], max_point, 'red', marker='o', markersize=5)
min_point = np.min(S)
plt.plot(t[np.argmin(S)], min_point, 'blue', marker='o', markersize=5)
meadle_point = (max_point + min_point)/2
# Настройки графика
plt.title('График изменения сигнала')
plt.xlabel('t')
plt.ylabel('S')
plt.grid()
plt.legend()
# Вывод графика
plt.show()
# Вывод максимального, минимального и среднего значений напряжения
print("Максимальное значение напряжения: ", max_point)
print("Минимальное значение напряжения: ", min_point)
print("Среднее значение напряжения: ", meadle_point)
```
